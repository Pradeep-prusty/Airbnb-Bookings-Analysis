# Taking Useful Columns Only

```python
# let's take usefull columns only and drop latitude,longitude,last_review,reviews_per_month columns
df = df.drop(["latitude","longitude","last_review","reviews_per_month"],axis=1)
df.head(5)
```

# **1. What can we learn about different hosts and areas?**

```python
hosts_areas = df.groupby(['host_name','neighbourhood_group']).agg({"calculated_host_listings_count":"max"}).sort_values(by='calculated_host_listings_count', ascending=False).reset_index()
hosts_areas.head(5)
```

**As we can see most number of listings are from Manhattan created Sonder (NYC)**

```python
# Number of Airbnb listed in each neighbourhood
neighbourhood =df.groupby(['neighbourhood_group']).agg({"id":"count"}).sort_values(by='id', ascending = False).reset_index()
neighbourhood.head()
```

```python
# Visualizing the number of listings in each neighbourhood
x = neighbourhood['neighbourhood_group']
y = neighbourhood['id']

fig = plt.figure(figsize = (10, 5))

plt.bar(x, y, width = 0.4,color ='red')
 
plt.xlabel("Neighbourhood Group")
plt.ylabel("Number Of Listing")
plt.title("Number of listings in neighbourhood group",{'fontsize':18})
plt.show()
```

# **2. What can we learn from predictions? (ex: locations, prices, reviews, etc)**

```python
areas_reviews = df.groupby(['neighbourhood_group']).agg({"number_of_reviews":"max"}).sort_values(by='number_of_reviews',ascending=False).reset_index()
areas_reviews.head()
```

```python
# Visualizing the number of reviews in neighbourhood_group
x = areas_reviews['neighbourhood_group']
y = areas_reviews['number_of_reviews']
fig = plt.figure(figsize = (10, 5))

plt.bar(x, y,width = 0.4,color ='red')
 
plt.xlabel("Groups")
plt.ylabel("Reviews")
plt.title("Groups & Reviews",{'fontsize':18})
plt.show()
```

```python
price_area = df.groupby(['price']).agg({"number_of_reviews":"max"}).reset_index()
price_area.head(5)
```

```python
#Visualizing the Reviews and Price
x = price_area['price']
y = price_area['number_of_reviews']
fig = plt.figure(figsize = (10, 5))
plt.stackplot(x, y, color='red')
plt.xlabel("Price")
plt.ylabel("Number of Review")
plt.title("Price & Number of Reviews",{'fontsize':18})
plt.show()
```

**From the above Analysis we can say that most people prefer to stay in place where price is less.**

# **3.Which hosts are the busiest and why?**

```python
busiest_hosts = df.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_hosts = busiest_hosts.sort_values(by='number_of_reviews', ascending=False).head(10)
busiest_hosts
```

```python
#Visualizing the host & number of reviews
x = busiest_hosts['host_name']
y = busiest_hosts['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
plt.bar(x, y,width = 0.4,color ='red')
plt.xlabel("Name of the Host")
plt.ylabel("Number of Reviews")
plt.title("Busiest Hosts",{'fontsize':18})
plt.show()
```

# **4. Is there any noticeable difference of traffic among different areas and what could be the reason for it?**

```python
traffic_areas = df.groupby(['room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by='minimum_nights')
traffic_areas
```

```python
# Visualizing the Room type and Night spend by peoples
x = traffic_areas['room_type']
y = traffic_areas['minimum_nights']
fig = plt.figure(figsize = (10, 10))
colors = ['#FF5F5F','#FF3333','#FF0000']
myexplode = [0,0.01,0.01] 
plt.pie (y , labels= x , startangle=45 ,colors= colors, autopct='%1.1f%%',explode = myexplode,shadow = True)
my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

**From the Above Analysis We can Stay that People are preferring Entire home/apt or Private room which are present in Manhattan, Brooklyn, Queens.**
